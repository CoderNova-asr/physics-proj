MB Dist: $n(\epsilon)= Aexp(-\beta\epsilon)=exp(-\beta(\epsilon-\mu)$

FD Dist:$n(\epsilon) = \frac{1}{\exp(\beta(\epsilon-\mu))+1}$

BE Dist:$n(\epsilon) = \frac{1}{\exp(\beta(\epsilon-\mu))1}$

```python
import numpy as np
import matplotlib.pyplot as plt
mu = 0
E = np.linspace(2e-10,2,1000)    #bE
```

```python
def mb(E):
    return 1/np.exp(E-mu)
```

```python
def fd(E):
    return 1/(np.exp(E-mu)+1)
```

```python
def be(E):
    return 1/(np.exp(E-mu)-1)
```

```python
plt.plot(E,mb(E))
plt.plot(E,fd(E))
plt.plot(E,be(E))

plt.legend(['Maxwell-Boltz','Fermi-Dirac','Bose-Einstein'])
plt.ylim(0,4)
```

Comparision for different temperatures

```python
def mb(E,T):
    m = 1/np.exp((E-mu)/k*T)
    return m
```

```python
k = 1.38e-23
n = int(input("Enter the number of temperatures for comparision: "))
T = np.zeros(n)
E = np.linspace(0,1e-24,1000)
mu = 0
print("Enter the temperature : ")
for i in range(n):
    T[i] = float(input())
```

```python
for i in range(n):
    plt.plot(E,mb(E,T[i]))
```
