```python
# -*- coding: utf-8 -*-
"""
Created on Wed Aug 25 21:14:15 2021

@author: Abhay Singh Rawat
19-17026
"""

import numpy as np
from numpy.linalg import eigh
import matplotlib.pyplot as plt


#taking standarrd values
m=940      #mass of proton
h=197.3
K= 100     #spring constant  

n=1000                      #steps
r=np.linspace(-5,5,n+1)     #all values of r from 0 to 1nm 
r=np.resize(r,n-1)          #sizing r so plotting is correct
ho=r[1]-r[0]            #step-size
#print("stepsize : ",ho)


#making the eigenvalue eq
k=h**2/(2*m)        

# since the dimensions of Hamiltonian shall be n*n
H=np.zeros([n-1,n-1])           
```

```python
print("Since the oscillator is anharmonic with:\n V(r)=(kr^2/2)+(br^3/3)")
x=int(input("how many values of  b you have: "))

for i in range(x):
    b=int(input("b :"))
    V=[]
    #making k+v ,i.e., hamiltonian matrix
    for i in range(n-1):
        for j in range(n-1):
            v=(K*r[i]*r[i]*0.5)+(b*r[i]**3)/3  #value of potential
            
            H[i,i]=2+v*ho*ho/k      #k+v in main diagonal and k is -2a
            if i==j+1:
                H[i,j]=-1        #lower diagonal
            if i==j-1:
                H[i,j]=-1           #upper diagonal  
        V.append(v)
    E,u = eigh(H)       #in-built for eigenvalues
    
    print('Ground state energy :',E[0]*k/(ho*ho),'MeV')
    
    plt.subplot(1,2,1)
    plt.plot(r,(np.absolute(u[:,0]))**2)
    plt.xlabel('distance from nucleus, r')
    plt.ylabel('probabilty density(|psi|^2)')
    
    
    plt.subplot(1,2,2)
    plt.plot(r,V,'r')
    plt.title('potential vs r')
    plt.show()
```
